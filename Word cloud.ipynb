# Find the frequency distribution

```python
from nltk.probability import FreqDist

fdist= FreqDist(words)

fdist.most_common(10)
```

- Notice that the most used words are punctuation marks and stopwords. We will have to remove such words to analyze the actual text.

# Plot the frequency graph

```python
import matplotlib.pyplot as plt

fdist.plot(10)
```

# Remove punctuation marks

```python
word_no_punc= []

for w in words:
    if w.isalpha():
        word_no_punc.append(w.lower())
        
print(word_no_punc)

print(len(word_no_punc))
```

```python
fdist= FreqDist(word_no_punc)

fdist.most_common(10)

fdist.plot(10)
```

- Notice that we still have many words that are not very useful in the analysis of our text file sample, such as and, but, so, and others. Next, we need to remove coordinating conjunctions.

# List of stopwords

```python
from nltk.corpus import stopwords

stopwords= stopwords.words("english")

print(stopwords)
```

```python
nltk.download('stopwords')
```

# Removing stopwords

```python
clean_words= []

for w in word_no_punc:
    if w not in stopwords:
        clean_words.append(w)
        
print(clean_words)

print(len(clean_words))
```

```python
fdist= FreqDist(clean_words)

fdist.most_common(10)

fdist.plot(10)
```

```python
import nltk
from nltk import word_tokenize
from nltk.corpus import stopwords
from wordcloud import WordCloud
import matplotlib.pyplot as plt

#Listing stop words
stopwords= stopwords.words("english")

def wordcloud(filename):
    
    
    text_file= open(filename)
    #Reading text

    text=text_file.read()

    #Converting into words
    words=word_tokenize(text)
    
    #Removing panctuations
    word_no_punc= []

    for w in words:
        if w.isalpha():
            word_no_punc.append(w.lower())
    

    
    #Removing stop words
    clean_words= []

    for w in word_no_punc:
        if w not in stopwords:
            clean_words.append(w)
            


    #Generating the word cloud
    word_cloud= WordCloud().generate(text)

    #Plot the word cloud
    plt.figure(figsize= (20,20))
    plt.imshow(word_cloud)

    #To remove the axis
    plt.axis("off")
```
